```python
import matplotlib.pyplot as plt
```

```python
plt.plot ([2,4, 6, 4])
plt.ylabel ("Number")
plt.xlabel ("Indices")
plt.title ("Myplot")
plt.show()
```

```python
plt.plot ([1, 2, 3, 4], [1, 4, 9, 16],)
plt.ylabel ("squares")
plt.xlabel ("numbers")
plt.grid()

plt.show()
```

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
plt.plot(x, y, linewidth=5.0)
plt.show()
```

```python
 import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline
from numpy.random import randn
```

```python
df.plot(kind = 'bar')
plt.show()
```
